```python
def proc_lic(s):
    global node
    id = node
    tree[id] = {"children":[],"meta":[]}
    node +=1
    c,m = s.pop(0),s.pop(0)
    for n in range(c):
        cn,s = proc_lic(s)
        tree[id]["children"].append(cn)
    for n in range(m):
        tree[id]["meta"].append(s.pop(0))
    return id,s
    
    
        
```

```python
def evaluate(node):
    v = 0
    c = tree[node]["children"]
    m = tree[node]["meta"]
    if not c:
        return sum(m)
    for n in m:
        if n - 1 < len(c):
            v += evaluate(c[n-1])
    return(v)
```
