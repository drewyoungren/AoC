```julia
function replace_track(c)
    if c in ('^','v') return '|'
    elseif c in ('<','>') return '-'
    else return c
    end
end

B = [replace_track(A[i,j]) for i in 1:150,j in 1:150];
```

```julia
DIR = Dict('<' => [0,-1],
    '>' => [0,1],'v' => [1,0],'^' => [-1,0]);

turnslashes = Dict('/'=>Dict('v'=>'<','^'=>'>','>'=>'^','<'=>'v'),
    '\\'=>Dict('v'=>'>','^'=>'<','>'=>'v','<'=>'^'))

turnplus = Dict('v'=>collect(">v<"),'^'=>collect("<^>"),
    '>'=>collect("^>v"),'<'=>collect("v<^"))
```

```julia
function tick(carts,A,B)
    newcarts = []
    for (x,y,c,n) in carts
#         println(x,y,c,n)
        nx,ny = [x,y] + DIR[c]
        d = A[nx,ny]
        A[x,y] = B[x,y]
        if d in ('-','|')
            nd  = c
            A[nx,ny] = nd
        elseif d in ('/','\\')
            nd = turnslashes[d][c]
            A[nx,ny] = nd
        elseif d == '+'
            n = mod1((n + 1) , 3)
            nd = turnplus[c][n]
            A[nx,ny] = nd
        else
            println("CRRRASH!! ",(nx,ny))
            A[nx,ny] = 'X'
            nd = c
#             break
        end
        push!(newcarts,[nx,ny,nd,n])
    end
    return sort(newcarts)
end
        
    
    
    
```
