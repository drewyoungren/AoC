```julia
work = Dict(v=>60 + k for (k,v) in enumerate('A':'Z'));
workers = [' ' for i in 1:5]
needs = Dict{Char,Set{Char}}(c => Set([]) for c in 'A':'Z')
for (k,v) in pairs(ikea)
    for st in v
#         needs[st] = get(needs,st,[])
        push!(needs[st],k)
    end
end

taken = []
clock = -1
while length(taken) < 26
    global clock
    # do work
    for (i,c) in enumerate(workers)
        if c in keys(work)
            work[c] += -1
            if work[c] < 1
                push!(taken,c)
                [delete!(v,c) for (k,v) in pairs(needs) if c in v]
                delete!(work,c)
                workers[i] = ' '
            end
        end
    end
#     println([(k,get(work,k,' ')) for k in workers])
    clock += 1
    available = findall(c -> c== ' ',workers)
    ready = sort([k for (k,v) in pairs(needs) if isempty(v) & !(k in vcat(taken,workers))])    
    slots = min(length(ready),length(available))
    for i in 1:slots
        g = popfirst!(ready)
        workers[available[i]] = g 
    end
    println(join([k for k in workers]))
end
```
