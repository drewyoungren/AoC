```python
def collapse(t):
    s = list(t)
    sout = ''
    store = ''
    while s:
        x = s.pop(0)
        if store == '' and s != []:
            store = x
        else:
            if store.lower() == x.lower() and store != x:
                store = ''
            else:
                sout += store
                store = x
#         print(sout)
    sout += store
    return sout
```

```python
def collapse2(t):
    pos = 0
    s = list(t)
    while pos < len(s) - 1:
        x = s[pos]
        y = s[pos+1]
        if x.lower() == y.lower() and x != y:
            s.pop(pos)
            s.pop(pos)
            pos = max(pos-1,0)
        else:
            pos += 1
    return ''.join(s)
```

```python
def oldcol(s):
    x = s
    y = collapse(x)
    while x != y:
        x = y
        y = collapse(x)
    return y
```
